```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load the dataset
credit_card_data = pd.read_csv("Testfraud.csv")

# Display some information about the dataset
print(credit_card_data.head())
print(credit_card_data.tail())
print(credit_card_data.info())
print(credit_card_data.isnull().sum())

# Check the distribution of fraud vs non-fraud cases
print(credit_card_data['is_fraud'].value_counts())

# Separate the data into legitimate and fraud cases
legit = credit_card_data[credit_card_data['is_fraud'] == 0]
fraud = credit_card_data[credit_card_data['is_fraud'] == 1]

# Display some statistics about the amounts in legitimate and fraud cases
print(legit['amt'].describe())
print(fraud['amt'].describe())

# Sample a subset of legitimate cases to balance the data
legit_sample = legit.sample(n=2145)

# Concatenate the sampled legitimate cases with the fraud cases
new_dataset = pd.concat([legit_sample, fraud], axis=0)

# Ensure 'is_fraud' column contains only numeric values
new_dataset['is_fraud'] = pd.to_numeric(new_dataset['is_fraud'], errors='coerce')

# Check the distribution of fraud vs non-fraud cases in the new dataset
print(new_dataset['is_fraud'].value_counts())

# Use groupby to calculate the mean of 'is_fraud'
print(new_dataset.groupby('is_fraud').mean())

# Check the data types
print(new_dataset.dtypes)

# Define features (x) and target (y)
x = new_dataset.drop(columns='is_fraud', axis=1)
y = new_dataset['is_fraud']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)

# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict on the training set
y_train_prediction = model.predict(X_train)

# Calculate training data accuracy
training_data_accuracy = accuracy_score(y_train_prediction, y_train)
print("Training Data Accuracy:", training_data_accuracy)
```
